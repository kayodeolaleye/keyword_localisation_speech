```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
path_word_freq = "data/word_freq_train.npy"
VOCAB_path = "data/flickr8k/transcript/VOCAB.pkl"
word_freq = np.load(path_word_freq)

with open(VOCAB_path, "rb") as f:
    vocab_to_id = pickle.load(f)

id_to_vocab = dict([(i[1], i[0]) for i in vocab_to_id.items()])
```

```python
vocab = [key for key in vocab_to_id]
```

```python
# Figure Size
fig, ax = plt.subplots(figsize =(16, 20))
 
# Horizontal Bar Plot
ax.barh(vocab, word_freq)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title('Word frequency in the training (audio) set by thresholding the soft labels with 0.4',
             loc ='left', )
 
# # Add Text watermark
# fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
#          color ='grey', ha ='right', va ='bottom',
#          alpha = 0.7)
 
# Show Plot
plt.show()
```
